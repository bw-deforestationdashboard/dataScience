```python
import pandas as pd

# Loading cleaned World Bank data on forest area as percent of land area...
pct_forest_coverage = pd.read_csv('pct_forest_coverage.csv', index_col=0)

pct_forest_coverage
```

```python
# Loading cleaned World Bank data on forest area in square kilometers...
sq_km_forest_coverage = pd.read_csv('sq_km_forest_coverage.csv', index_col=0)

sq_km_forest_coverage
```

```python
# Putting time series data for countries in tidy format...
pct_forest_coverage_tidy = pct_forest_coverage.melt(
    id_vars=['Country', 'Code'])
sq_km_forest_coverage_tidy = sq_km_forest_coverage.melt(
    id_vars=['Country', 'Code'])

# Renaming columns...
pct_forest_coverage_tidy = pct_forest_coverage_tidy.rename(
    columns={'variable': 'Year', 'value': 'Coverage'})
sq_km_forest_coverage_tidy = sq_km_forest_coverage_tidy.rename(
    columns={'variable': 'Year', 'value': 'Coverage'})

# Sorting by country...
pct_forest_coverage_tidy = pct_forest_coverage_tidy.sort_values(
    by=['Country'])
sq_km_forest_coverage_tidy = sq_km_forest_coverage_tidy.sort_values(
    by=['Country'])

# Resetting indices...
pct_forest_coverage_tidy = pct_forest_coverage_tidy.reset_index()
sq_km_forest_coverage_tidy = sq_km_forest_coverage_tidy.reset_index()

# Dropping rows with nulls...
pct_forest_coverage_tidy = pct_forest_coverage_tidy[
    pct_forest_coverage_tidy['Coverage'].notnull()]
sq_km_forest_coverage_tidy = sq_km_forest_coverage_tidy[
    sq_km_forest_coverage_tidy['Coverage'].notnull()]
```

```python
from sklearn.linear_model import LinearRegression

# Preparing linear regression to make predictions for each country...
model = LinearRegression()

# Getting list of country codes...
codes = pct_forest_coverage_tidy['Code'].unique()

# Getting list of years where we will predict forest coverage...
years = [year for year in range(2017, 2025)]

# For-loop to make predictions for each country with first dataset...
rows = []
for code in codes:
    dictionary = {'Code': code}
    model.fit(pct_forest_coverage_tidy[
        pct_forest_coverage_tidy['Code'] == code][['Year']],
              pct_forest_coverage_tidy[
        pct_forest_coverage_tidy['Code'] == code]['Coverage'])
    for year in years:
        prediction = model.predict([[year]])
        dictionary[str(year)] = prediction[0]
    rows.append(dictionary)

# Making a new dataframe out of the predictions for the first dataset...
pct_forest_coverage_predictions = pd.DataFrame(rows)

# Reordering the columns in the dataframe for the first dataset...
pct_forest_coverage_predictions = pct_forest_coverage_predictions[
    ['Code'] + [str(year) for year in years]]

pct_forest_coverage_predictions
```

```python
# For-loop to make predictions for each country with second dataset...
rows = []
for code in codes:
    dictionary = {'Code': code}
    model.fit(sq_km_forest_coverage_tidy[
        sq_km_forest_coverage_tidy['Code'] == code][['Year']],
              sq_km_forest_coverage_tidy[
        sq_km_forest_coverage_tidy['Code'] == code]['Coverage'])
    for year in years:
        prediction = model.predict([[year]])
        dictionary[str(year)] = prediction[0]
    rows.append(dictionary)

# Making a new dataframe out of the predictions for the first dataset...
sq_km_forest_coverage_predictions = pd.DataFrame(rows)

# Reordering the columns in the dataframe for the first dataset...
sq_km_forest_coverage_predictions = sq_km_forest_coverage_predictions[
    ['Code'] + [str(year) for year in years]]

sq_km_forest_coverage_predictions
```

```python
import numpy as np

# Getting list of country codes where we have predictions...
pct_forest_coverage_prediction_codes = [
    code for code in pct_forest_coverage_predictions['Code']]
sq_km_forest_coverage_prediction_codes = [
    code for code in sq_km_forest_coverage_predictions['Code']]

# Getting full list of country codes...
all_codes = [code for code in pct_forest_coverage['Code']]

# Finding country codes where we have no predictions because all data null...
pct_forest_coverage_prediction_nulls = [
    item for item in all_codes if item 
    not in pct_forest_coverage_prediction_codes]
sq_km_forest_coverage_prediction_nulls = [
    item for item in all_codes if item 
    not in sq_km_forest_coverage_prediction_codes]

# Creating null prediction data for first dataset...
rows = []
for code in pct_forest_coverage_prediction_nulls:
    dictionary = {'Code': code}
    for year in years:
        dictionary[str(year)] = np.nan
    rows.append(dictionary)

# Turning null prediction data for first dataset into a dataframe...
pct_forest_coverage_null_pred_df = pd.DataFrame(rows)

# Reordering the columns in the null dataframe...
pct_forest_coverage_null_pred_df = pct_forest_coverage_null_pred_df[
    ['Code'] + [str(year) for year in years]]

pct_forest_coverage_null_pred_df
```

```python
# Creating null prediction data for second dataset...
rows = []
for code in sq_km_forest_coverage_prediction_nulls:
    dictionary = {'Code': code}
    for year in years:
        dictionary[str(year)] = np.nan
    rows.append(dictionary)
    
# Turning null prediction data for first dataset into a dataframe...
sq_km_forest_coverage_null_pred_df = pd.DataFrame(rows)

# Reordering the columns in the null dataframe...
sq_km_forest_coverage_null_pred_df = sq_km_forest_coverage_null_pred_df[
    ['Code'] + [str(year) for year in years]]

sq_km_forest_coverage_null_pred_df
```

```python
# Appending null dataframe to the predictions dataframe for first dataset...
pct_forest_coverage_predictions = pct_forest_coverage_predictions.append(
    pct_forest_coverage_null_pred_df)

# Getting country names for the predictions dataframe for first dataset...
pct_forest_coverage_predictions = pct_forest_coverage_predictions.merge(
    pct_forest_coverage[['Country', 'Code']], on='Code')

# Reordering columns in the dataframe...
pct_forest_coverage_predictions = pct_forest_coverage_predictions[
    ['Country', 'Code'] + [str(year) for year in years]]

# Sorting by country name...
pct_forest_coverage_predictions = pct_forest_coverage_predictions.sort_values(
    by=['Country'])

# Resetting index...
pct_forest_coverage_predictions = pct_forest_coverage_predictions.reset_index(
    drop=True)

pct_forest_coverage_predictions
```

```python
# Appending null dataframe to the predictions dataframe for first dataset...
sq_km_forest_coverage_predictions = sq_km_forest_coverage_predictions.append(
    sq_km_forest_coverage_null_pred_df)

# Getting country names for the predictions dataframe for first dataset...
sq_km_forest_coverage_predictions = sq_km_forest_coverage_predictions.merge(
    pct_forest_coverage[['Country', 'Code']], on='Code')

# Reordering columns in the dataframe...
sq_km_forest_coverage_predictions = sq_km_forest_coverage_predictions[
    ['Country', 'Code'] + [str(year) for year in years]]

# Sorting by country name...
sq_km_forest_coverage_predictions = (
    sq_km_forest_coverage_predictions.sort_values(by=['Country']))

# Resetting index...
sq_km_forest_coverage_predictions = (
    sq_km_forest_coverage_predictions.reset_index(drop=True))

sq_km_forest_coverage_predictions
```

```python
# Exporting dataframe as csvs and jsons...

pct_forest_coverage_predictions.to_csv(
    'pct_forest_coverage_predictions.csv')
pct_forest_coverage_predictions.to_json(
    'pct_forest_coverage_predictions.json', orient='records')

sq_km_forest_coverage_predictions.to_csv(
    'sq_km_forest_coverage_predictions.csv')
sq_km_forest_coverage_predictions.to_json(
    'sq_km_forest_coverage_predictions.json', orient='records')
```
