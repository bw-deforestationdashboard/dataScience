```python
import pandas as pd

# Loading World Bank data on forest area as percent of land area...
pct_forest_coverage = pd.read_csv(
    'Raw_World_Bank_Data/e3f7f706-7166-42a3-ae92-a9989f141bd8_Data.csv')

pct_forest_coverage
```

```python
# Loading World Bank data on forest area in square kilometers...
sq_km_forest_coverage = pd.read_csv(
    'Raw_World_Bank_Data/131f20da-d42b-4bfc-a9be-47ef34d8cc78_Data.csv')

sq_km_forest_coverage
```

```python
import geopandas as gpd

# Loading Thematic Mapping data on country boundaries...
country_boundaries = gpd.read_file(
    'Raw_Thematic_Mapping_Data/TM_WORLD_BORDERS-0.3/TM_WORLD_BORDERS-0.3.shp')

country_boundaries
```

```python
import numpy as np

# Dropping rows from World Bank data where boundary data is lacking...
pct_forest_coverage = pct_forest_coverage[
    pct_forest_coverage['Country Code'].isin(country_boundaries['ISO3'])]
sq_km_forest_coverage = sq_km_forest_coverage[
    sq_km_forest_coverage['Country Code'].isin(country_boundaries['ISO3'])]

# Resetting indices...
pct_forest_coverage = pct_forest_coverage.reset_index(drop=True)
sq_km_forest_coverage = sq_km_forest_coverage.reset_index(drop=True)

# Dropping columns identifying dataset since value is always the same...
pct_forest_coverage = pct_forest_coverage.drop(
    columns=['Series Name', 'Series Code'])
sq_km_forest_coverage = sq_km_forest_coverage.drop(
    columns=['Series Name', 'Series Code'])

# Simplifying remaining column names...
years = [year for year in range(1990, 2017)]
columns = ['Country', 'Code'] + years
sq_km_forest_coverage.columns = columns
pct_forest_coverage.columns = columns

# Converting string representations of null values to numpy nans...
pct_forest_coverage = pct_forest_coverage.replace('..', np.nan)
sq_km_forest_coverage = sq_km_forest_coverage.replace('..', np.nan)

# Changing data types to make numeric features numeric...
pct_forest_coverage[years] = pct_forest_coverage[years].astype(float)
sq_km_forest_coverage[years] = sq_km_forest_coverage[years].astype(float)
```

```python
# Checking first World Bank dataset...

pct_forest_coverage
```

```python
# Checking second World Bank dataset...

sq_km_forest_coverage
```

```python
# Dropping rows from Thematic Mapping data where forest data is lacking...
country_boundaries = country_boundaries[
    country_boundaries['ISO3'].isin(pct_forest_coverage['Code'])]

# Resetting index...
country_boundaries = country_boundaries.reset_index(drop=True)

# Dropping all columns except country code and boundary data...
country_boundaries = country_boundaries[['ISO3', 'geometry']]

# Making country code column name consistent with World Bank datasets...
country_boundaries = country_boundaries.rename(columns={'ISO3': 'Code'})

# Checking Thematic Mapping dataset...
country_boundaries
```

```python
# Exporting dataframes as csvs and jsons/geojsons...

pct_forest_coverage.to_csv('pct_forest_coverage.csv')
pct_forest_coverage.to_json('pct_forest_coverage.json', orient='records')

sq_km_forest_coverage.to_csv('sq_km_forest_coverage.csv')
sq_km_forest_coverage.to_json('sq_km_forest_coverage.json', orient='records')

country_boundaries.to_csv('country_boundaries.csv')
country_boundaries.to_file("country_boundaries.geojson", driver='GeoJSON')
```
