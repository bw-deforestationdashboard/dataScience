```python
import pandas as pd

# Loading cleaned World Bank data on forest area as percent of land area...
pct_forest_coverage = pd.read_csv('pct_forest_coverage.csv', index_col=0)

pct_forest_coverage
```

```python
# Loading predicted data on forest area in square kilometers...
pct_forest_coverage_predictions = pd.read_csv(
    'pct_forest_coverage_predictions.csv', index_col=0)

pct_forest_coverage_predictions
```

```python
import geopandas as gpd

# Loading cleaned Thematic Mapping data on country boundaries...
country_boundaries = gpd.read_file('country_boundaries.geojson')

country_boundaries
```

```python
import folium

# Passing folium country boundary data to use for mapping...
countries = folium.features.GeoJson(country_boundaries)

# Creating max number of equally sized bins for choropleth; max is nine...
bins = [(11*n) for n in range(10)]

# Creating list of years where World Bank data is available...
years = [str(year) for year in range(1990, 2017)]

# For-loop to create a choropleth map for each year...
for year in years:
    # Generating world map at appropriate zoom...
    folium_map = folium.Map(
        location=[40, 10], zoom_start=2, tiles='cartodbpositron')
    # Adding country boundary data to generated map...
    folium_map.add_child(countries)
    # Making map a choropleth with World Bank data on forest coverage...
    folium_map.choropleth(
        geo_data=country_boundaries,
        name='choropleth',
        data=pct_forest_coverage,
        columns=['Code', year],
        key_on='feature.properties.Code',
        fill_color='Greens',
        fill_opacity=1,
        line_opacity=.1,
        legend_name=('Forest Area as % of Land Area in ' + year),
        bins=bins,
        reset=True
    )
    folium.LayerControl().add_to(folium_map)
    
    file_name = ('Folium_Maps/html/pct_forest_coverage_' + year + '_map.html')
    
    folium_map.save(file_name)
```

```python
# Creating list of years where we made predictions...
years = [str(year) for year in range(2017, 2025)]

# For-loop to create a choropleth map for each predicted year...
for year in years:
    # Generating world map at appropriate zoom...
    folium_map = folium.Map(
        location=[40, 10], zoom_start=2, tiles='cartodbpositron')
    # Adding country boundary data to generated map...
    folium_map.add_child(countries)
    # Making map a choropleth with World Bank data on forest coverage...
    folium_map.choropleth(
        geo_data=country_boundaries,
        name='choropleth',
        data=pct_forest_coverage_predictions,
        columns=['Code', year],
        key_on='feature.properties.Code',
        fill_color='Greys',
        fill_opacity=1,
        line_opacity=.1,
        legend_name=('Predicted Forest Area as % of Land Area for ' + year),
        bins=bins,
        reset=True
    )
    folium.LayerControl().add_to(folium_map)
    
    file_name = ('Folium_Maps/html/pct_forest_coverage_prediction_'
                 + year + '_map.html')
    
    folium_map.save(file_name)
```
